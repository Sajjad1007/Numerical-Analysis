```python
print("The given equation, x^2 - 6x + 7 = 0\n")

x = 5   # initial guess

for i in range(1000):
    FofX = x**2 - 6*x + 7
    F_ofX = 2*x - 6
    xnew = x - (FofX / F_ofX)
    error = (abs(xnew - x) / xnew) * 100
    print("Iteration no %d :" %(i + 1), "Error = %0.6g%%" %error)
    
    if abs(xnew - x) <= 0.000001:
        x = xnew
        break
    else:
        x = xnew
        
print("\nThe root = %0.6g" %x)
print("Number of iterations :", i + 1)
```

```python
import sys
from math import sqrt

print("The given equation, x^2 - 6x + 7 = 0\n")

x1 = 0   # initial guess

for i in range(1000):
    xnew = (x1**2 + 7) / 6
    
    if abs(xnew - x1) <= 0.000001:
        x1 = xnew
        break
    else:
        x1 = xnew

x2 = 5   # initial guess

for j in range(1000):
    d = 6*x2 - 7        
    xnew =  sqrt(d)
    
    if abs(xnew - x2) <= 0.000001:
        x2 = xnew
        break
    else:
        x2 = xnew
        
print("The roots : x1 = %0.6g," %x1, "x2 = %0.6g" %x2)
print("Number of iterations : for x1 = %d" %(i + 1), "and for x2 = %d" %(j + 1))
```
