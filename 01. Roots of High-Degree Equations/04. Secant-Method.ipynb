```python
def f(x):
    y = x**2 - 4*x - 5
    return y


def secant(x1, x2):
    n = int(input("Maximum iterations : "))
    
    for i in range(n):
        y1 = f(x1)
        y2 = f(x2)
        xnew = x2 - y2 * ((x2 - x1) / (y2 - y1))
        
        if abs(xnew - x2) < 0.000001:
            print("\nThe root = %0.5f" %xnew, "(%d iterations)" %(i + 1))
            return
        else:
            x1 = x2
            x2 = xnew
        
    print("\nThe root = %0.5f" %xnew, "(%d iterations)" %(i + 1))
    return
    

# Main code starts from here

print("The given equation, x^2 - 4x - 5 = 0\n")

x1 = float(input("Enter 1st initial guess = "))
x2 = float(input("Enter 2nd initial guess = "))
secant(x1, x2)
```
